# **What is the Runge-Kutta Method?**

First developed by Carl Runge (1856 to 1927) and Marin Kutta (1867 to 1944) in order to study atomic spectra, the Runge-Kutta method is a method of numerically integrating ordinary differential equations (ODE’s) by using a trial step at the midpoint of an interval to cancel our lower-order error terms. In other words given an ODE and an initial condition, the result is an estimated slope of the ODE around a fixed point. 

$y' = f(x,y)$

$y(x_0) = y_0$

*Given an ODE and an initial condition, we can estimate the slope around a fixed point*

In order to effectively teach differentials however, we had to restrict ourselves to specifics types of solvable first-order ODE's – that is linear, separable, and exact to name a few. However, most first-order ODE's are in fact not as easily solvable (or even solvable at all, for that matter) with the methods we currently have. In order to estimate the result of an ODE given an initial point, we can utilize the Runge-Kutta Method. The Runge-Kutta method itself consists of several other sub-types of methods. The most common include the first-order, second-order, and fourth-order Runge-Kutta Methods.

For the most part, each sub-type of the Runge-Kutta Method is conceptually the same. Each Runge-Kutta method is a single-step method, but with multiple stages per step. Given an ODE and an initial condition, we must first "chop" the initial condition into several smaller subdivisions of length *h*. How many smaller subdivisions ultimately depends on the type of the Runge-Kutta method. The solution is then generated using a series of iterative formulas that are unique to each Runge-Kutta sub-type method. We continue to use these iterative formuals until a constant approximated solution is found. Each method is outlined more in-depth below.



# **First-Order Runge-Kutta Method**

First published in 1768 by Leohard Euler in his three-volume textbook *Institutionum calculi integralis*, the First-Order Runge-Kutta Method is also more widely conventionally known as Euler's Method. Although there does exist both a Euler Forward and Backward Method, we will only discuss the Euler's Foward Method (oftentimes shortened as simply just Euler's Method) for the purposes of simplicification. The only significant difference between each method is that Euler's Backward Method allows for larger time steps to be taken. Euler's Method's iterative formula however is outlined below.

$y_{n+1} = y_n + h(x_n, y_n)$

The issue with Euler's Method is that it is the most accurate with functions that do not change incredibly rapidly. Additionally, we must ensure that our step-size, that is our *h* value, isn't too large or our numerical estimation will also be inaccurate. In order to work around this issue, larger order Runge-Kutta methods are used. This is not to suggest however that there exists a direct relationship between higher-order Runge-Kutta Methods and numerical approximation accuracy. In other words: a higher order Runge-Kutta method does not always suggest a greater approximation compared to a lower order Runge-Kutta method

# **Second-Order Runge-Kutta Method**

Conceptually speaking, the same ideas used in the First-Order Runge-Kutta Method are usedly similarly in the Second-Order Runge-Kutta Method. Unlike Euler's Method, there does not exist a forward or backward method. Rather, the Second-Order Runge-Kutta Method uses a time step at the midpoint of an interval to cancel out lower-order error terms. It's formula is outlined below.

$k_1 = hf(x_n, y_n)$

$k_2 = hf(x_n+\frac12h, y_n+\frac12k_1)$

$y_{n+1} = y_n + \frac{k_1 + k_2}2$

Notice that unlike Euler's Method, there exists both $k_1$ and $k_2$. They are conventionally called the stages of the Runge-Kutta Method. Ultimately, the number of stages corresponds to the order of the Runge-Kutta method. Each additional step corresponds to a different estimation of the slope as well as greater time-steps or $h$ values used. Thus there exists some correlation between the greater number of stages and the accuracy in estimating the ODE.

# **Fourth-Order Runge-Kutta Method**

Much unlike both Euler's Method and the Second-Order Runge-Kutta Method, the Fourth-Order Runge-Kutta Method uses four stages. It's iterative formulas are outlined below.

$k_1 = hf(x_n, y_n)$

$k_2 = hf(x_n + \frac12h, y_n + \frac12k_1)$

$k_3 = hf(x_n + \frac12h, y_n + \frac12k_2)$

$k_4 = hf(x_n + h, y_n + k_3)$

$y_{n+1} = y_n + \frac16k_1 + \frac13k_2 + \frac13k_3 + \frac16k_4 + O(h^5)$

Ultimately out of all the discussed Runge-Kutta Methods, the Fourth-Order Runge-Kutta Method is the most accurate because of the greater number of evaluations used. Although more conventionally used in approximating initial value problems, a variation of the Fourh-Order Runge-Kutta method exists and is briefly discussed below

# **3/8 Fourth-Order Runge-Kutta Method**

Rather than increasing the order of the Runge-Kutta method to obtain a different estimation, we can also vary the slopes while continuing to keep the four times per step evaluations. In fact, the 3/8 Fourth-Order Runge-Kutta Method (abbreivated as simply just the 3/8 Runge-Kutta Method) achieves this and is outlined below.

$k_1 = hf(x_i, y_i)$

$k_2 = hf(\frac{x_i+h}3, \frac{y_i+k_1}3)$

$k_3 = hf(\frac{x_i + 2h}3, \frac{y_i-k_1}3 + k_2)$

$k_4 = hf(x_i + h, y_i + k_1 - k_2 + k_3)$

$y_{i+1} = y_i + \frac18(k_1 + 3k_2 + 3k_3 + k_4)$

Unlike the more conventionally known Fourth-Order Runge Kutta Method which uses averages of $\frac16$, $\frac13$, $\frac13$, and $\frac16$, the 3/8 Runge-Kutta Method uses averages of $\frac18$, $\frac38$, $\frac38$, and $\frac18$. A different set of weighted averages, however, does not indicate a better estimation. Depending on the initial value problem itself, the 3/8 Fourth-Order Runge-Kutta Method may obtain a better estimation than the conventional Fourth-Order Runge-Kutta method. More limitations of each the Fourth-Order Runge-Kutta Method are also discussed below.

# **Limitations and Accuracy Analysis of Fourth-Order Runge-Kutta Methods**

Although the Fourth-Order Runge-Kutta method is the most widely used as well as the most accurate, it has several limitations.

For one thing, physically computing it is demanding in itself because of the greater number of time steps used. In other words, handwriting solutions to initial value problems becomes tedious. With that said however, progammatically coding a simple function in any coding language offers a viable workaround. In fact, hard-coding a function can be as short as ten lines of code in most languages.

Although the standard Fourth-Order Runge-Kutta Method is the most widely used, it does not guarantee the most accurate approximation for *all* initial value problems. In fact, some initial value problems can have better approximations obtained using the alternative 3/8 Fourth-Order Runge-Kutta method. A comparison analysis is performed below.
